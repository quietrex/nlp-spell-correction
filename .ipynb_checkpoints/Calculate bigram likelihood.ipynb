```python
def get_stemmed_word(bigram):
        tokens = nlp(bigram)
        list_stemmed_word = ""
        for token in tokens:
            list_stemmed_word += token.lemma_ + " "
        return(list_stemmed_word.strip())
```

```python
def check_realword_error(bigram):
    j = []
    for note in bigram:
        content = note[0] + ' ' + note[1]
        stemmed_word = get_stemmed_word(content)
        
        if get_bigram_score(content) == 0:
            # real word error
            j.append(False)
        else:
            if get_bigram_score(stemmed_word) == 0 or get_bigram_score(stemmed_word) == get_bigram_score(content):
                j.append(True)
            elif get_bigram_score(stemmed_word) > get_bigram_score(content):
                j.append(False)
                
    return j
```

```python
def edit_distance(str1, str2, m, n):
    dp = [[0 for x in range(n + 1)] for x in range(m + 1)]
    
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                dp[i][j] = j
            elif j == 0:
                dp[i][j] = i
            elif str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i][j-1],
                                   dp[i-1][j],
                                   dp[i-1][j-1])
    return dp[m][n]
```

```python
import numpy as np
def suggest_word(word1, word2, selection):
    def get_suggestion_List(word1, word2, selection):
        bigrams_text = json.load(open("training_data\\bigrams_train_set.txt"))
        if selection == 1:
            bigram_suggestion_list_1 = []
            for i in bigrams_text:
                if i[1] == word2 and i not in bigram_suggestion_list_1:
                    bigram_suggestion_list_1.append(i)
            return bigram_suggestion_list_1
        
        else:
            bigram_suggestion_list_2 = []
            for i in bigrams_text:
                if i[0] == word1 and i not in bigram_suggestion_list_2:
                    bigram_suggestion_list_2.append(i)
            return bigram_suggestion_list_2
        
    score_list = {}
    for i in get_suggestion_List(word1, word2, selection):
        bigram_wrd = i[0] + " " + i[1]
        edit_distanc_1 = edit_distance(word1, i[0], len(word1), len(i[0]))
        edit_distanc_2 = edit_distance(word2, i[1], len(word2), len(i[1]))
        if selection == 1:
            if edit_distanc_1 < 10:
                j = str(np.around(get_bigram_score(bigram_wrd), 4)) + " [" + str(edit_distanc_1) + "]"
                score_list[bigram_wrd] = edit_distanc_1
        else:
            if edit_distanc_2 < 10:
                j = str(np.around(get_bigram_score(bigram_wrd), 4)) + " [" + str(edit_distanc_2) + "]"
                score_list[bigram_wrd] = edit_distanc_2
                
    return score_list

suggestion_1 = suggest_word('over','four',1)
suggestion_2 = suggest_word('over','four',2)
```
